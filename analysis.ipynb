```python
# Viewing data
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("survey_data.csv")

df.head()
```

```python
# Age with highest income
age_income = df.groupby("Age")["Income"].mean().sort_values(ascending=False)

plt.figure()
age_income.plot(kind="bar")
plt.xlabel("Age")
plt.ylabel("Average Income")
plt.title("Average Income by Age")
plt.tight_layout()
plt.savefig("income_by_age.png")
plt.show()
```

```python
# Reshaping the data for expense analysis
expense_cols = [
    "Utilities",
    "Entertainment",
    "School Fees",
    "Shopping",
    "Healthcare"
]

gender_expenses = df.groupby("Gender")[expense_cols].sum()
```

```python
# Plotting spending by gender
gender_expenses.plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Total Spending")
plt.title("Spending by Gender Across Categories")
plt.tight_layout()
plt.savefig("spending_by_gender.png")
plt.show()
```
